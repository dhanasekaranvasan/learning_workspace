```python
# list, tuple, set, frozenset, dict
nums = [50, 48, 32, 45, 68, 76, 48]
print("---List---")
for num in nums:
    print(num, end=", ")

names = {"Arun", "Kishore", "Selva", "Venky", "Vasanth"}
print("\n---Set---")
for name in names:
    print(name, end=", ")

mixed = (3, True, "Arun", "Hello world", 7.1, 234, 89)
print("\n---Tuple---")
for item in mixed:
    print(item, end=", ")

items = {"Name": "Dhana", "Age": 32, "Height": 6.2, "Weight": 70, "Technology": ["Java", "Python", "JavaScript"]}
print("\n---Dict---")
for key, value in items.items():
    print(f'{key}: {value}', end=", ")
```

```python
# break and continue
# break -> exit from current running loop
# continue -> go to next iteration, if continue work skip every logic below 'continue'
nums = [50, 48, 32, 45, 68, 76, 48]
print("break")
for num in nums:
    if num == 68:
        break
    print(num, end=" ")
print("\ncontinue")
for num in nums:
    if num == 68:
        continue
    print(num, end=" ")

for i in range(5): # loop 1
    for j in range(5): # loop 2
        if i == 2 and j == 2:
            break # exit from loop 2
        print("Loop 2: ", i, j)
    if i == 3:
        break # exit from loop 1
    print("Loop 1: ", i)

for i in range(5): # loop 1
    for j in range(5): # loop 2
        if i == 2 and j == 2:
            continue # go to next iteration in loop 2
        print("Loop 2: ", i, j)
    if i == 3:
        continue # go to next iteration in loop 1
    print("Loop 1: ", i)
```

```python
# else block (execute when without trigger break)
for num in range(1, 6):
    print(num)
else:
    print("Loop executed successfully without break")

for num in range(1, 11):
    if num == 20: # condition not success so no break
        break 
    print(num)
else:
    print("Loop executed successfully without break")

for num in range(1, 11):
    if num == 7: # condition true success so no break
        break 
    print(num)
else:
    print("Loop executed successfully without break")
```

```python
# while loop (loop run until condition fails)
count = 0
while True:
    if count == 15:
        break
    print(count)
    count += 1

print()
count = 11
while count >= 0:
    print(count)
    count -= 1
```
