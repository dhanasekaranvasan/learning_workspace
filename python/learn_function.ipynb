```python
# create function using 'def' keyword
def firstFunc():
    pass # nothing keyword just used for syntax
firstFunc()
```

```python
# create function with parameter
def sample(name):
    print(f'Name: {name}')
sample('Josh') # positional argument
sample(name='Josh') # keyword based argument
```

```python
# create function with default argument
def sample(name, age=32):
    print(f'Name: {name}, Age: {age}')
sample('Josh') # same as sample('Josh', 32)
sample('Josh', 30)
sample(age=32, name='Josh')
sample(name='Josh')
```

```python
# create dynamic positional arguments (*args -> tuple) 
def customer(*args):
    print(f'Name: {args[0]}, Age: {args[1]}, Height: {args[2]}, Weight: {args[3]}')
customer('Allen', 32, 5.9, 65)

def customer2(name, *args):
    print(f'Name: {name}, Age: {args[0]}, Height: {args[1]}, Weight: {args[2]}')
customer2('Allen', 27, 6.0, 70)
```

```python
# create dynamic keyword arguments (**kwargs -> dict)
def customer(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}: {value}')
customer(name='Allen', age=32, height=6.2, weight=173)
def customers(name, **kwargs):
    print(f'Name: {name}')
    for key, value in kwargs.items():
        print(f'{key}: {value}')
customers('Allen', age=32, height=6.2, weight=173)
```

```python
# create dynamic positional and keyword arguments (*args, **kwargs)
# *args -> tuple()
# **kwargs -> dict
def info(*args, **kwargs):
    print(args)
    print(kwargs)
info('Hello area', 51, '06-07-1879', 'America', name='Allen', age=32, height=6.2, weight=78)
```

```python
# global and nonlocal variable
x = 10 # gloabl
def number():
    x = 20 # local
number()
print(x) # global variable

x = 30
def number2():
    global x
    x = 40 # global variable
number2()
print(x)

x = 10 # gloabl variable
def outer():
    x = 20 # local variable
    def inner():
        global x # gloabl variable
        x = 50
    inner()
    print(x) # outer local variable
outer()
print(x)

y = 10
def outer():
    y = 20
    def inner():
        nonlocal y
        y = 30
    inner()
    print(y)
outer()
print(y)
```

```python
# return function
# return ->  (default return is 'None')
    # return types -> int, str, bool, list, tuple, dict, set, frozenset, object 
# syntax: def function_name():
# fixed return type: def function_name() -> <return type>:
def sample():
    return 1
print(sample())
def main(func):
    return func
def func():
    print("this is function call")
call = main(func)
call()
def returntype() -> list:
    return [num for num in range(1, 11)]
print(returntype())
```

```python
# recursion function
def reverse(num):
    if (num<1):
        return
    print('Number: ', num)
    return reverse(num - 1)
sample(10)
```
