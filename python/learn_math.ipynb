```python
import math

print('Power of 2:', pow(2, 4))
print('Pwer of 3:', pow(3, 3))

print('Absolute value of -18:', abs(-18))
print('Absolute value of -9:', abs(-9))

print('Square root of 40:', math.sqrt(36))
print('Square root of 57:', math.sqrt(57))

pi = round(math.pi, 2)
r = 10.5
print('Circumference of circle:', 2 * pi * r)
print('Area of circle:', pi * pow(r, 2))
print('Diameter of circle:', 2 * r)

print('Ceil round the value to next (.1 to .9):', math.ceil(4.345))
print('Floor round the value to same number (.1 to .9):', math.floor(7.119))
```
