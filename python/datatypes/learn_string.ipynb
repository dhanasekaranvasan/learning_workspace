```python
# casefold(Unicode character) remove all the cases from the string (Internationalization)
txt = 'Hello World!'
print(txt)
print(txt.casefold())
text = 'groß'
print(text.casefold())
```

```python
# center the string
# str.center(width: SupportsIndex, fillChar: str = ' ') -> str
# width: SupportsIndex -> int, def, bool -> True = 1, False = 0
# fillChar: str -> single character only (default is space(' '))
txt = 'hello world!'
print(txt.center(True)) # len(txt) => 12, width => 1, fillChar => len(txt) - 1, 12 - 1 => 1 < 12 no fillchar
print(txt.center(33)) # len(txt) => 12, width => 33, fillChar => len(txt) - width, 12 - 33 => 33 > 12
print(txt.center(40, '1')) # len(txt) => 12, width => 40, fillChar => len(txt) - width, 12 - 40 => 40 > 12
print(txt.center(15, '*')) # len(txt) => 12, width => 15, fillChar => len(txt) - width, 12 - 15 => 15 > 12
print(txt.center(20, '&')) # len(txt) => 12, width => 20, fillChar => len(txt) - width, 12 - 20 => 20 > 12
```

```python
# count the single or more lettes
# str.count(sub: str, start: SupportsIndex | None = None, end: SupportsIndex | None = None,) -> int
# sub: str -> single or more characters
# start: SupportsIndex -> int, def, bool -> True = 1, False = 0
# end: SupportsIndex -> int, def, bool -> True = 1, False = 0
txt = 'hello world!'
print(txt.count('l')) # total count in the string
print(txt.count('hello')) #total count in the string
print(txt.count('llo', 2)) # total count in the string from index 2
print(txt.count('h', False, True)) # total count in the string from index 0 to 1 (1st excluded)
print(txt.count('l', 5, 11)) # total count in the string from index 5 to 10 (11th excluded)
print(txt.count('l', 5, -3)) # total count in the string from between index 5 to last 3rd (3rd included)
```

```python
# endswith character or characters
# str.endswith(suffix: str | tuple[str, ...], start: SupportsIndex | None = None, end: SupportsIndex | None = None,) -> bool
# suffix: str | tuple[str, ...] -> single or more characters or tuple()
# start: SupportsIndex -> int, def, bool -> True = 1, False = 0
# end: SupportsIndex -> int, def, bool -> True = 1, False = 0
txt = 'hello world!'
print(txt.endswith('!', False)) # ends with single chacater
print(txt.endswith('world!')) # ends with more characters
print(txt.endswith('w', 0, -5)) # ends with character from index 0 to last 4th (last 3rd excluded)
print(txt.endswith('worl', 0, -4)) # ends with more characters from index 0 to last 5th (last 2nd excluded)
# same as string with one or more string
print(txt.endswith(('!')))
print(txt.endswith(('d', 'o'), 0, -1))
```

```python
# find the index of one or more characters of first occurance (-1 if no occurance)
# str.find(sub: str, start: SupportsIndex | None = None, end: SupportsIndex | None = None,) -> int
# sub: str
# start: SupportsIndex -> int, def, bool -> True = 1, False = 0
# end: SupportsIndex -> int, def, bool -> True = 1, False = 0
txt = 'hello world!'
print(txt.find('hello')) # first occurance of characters
print(txt.find('l', True)) # first occurance of character from index 1
print(txt.find('o w')) # first occurance of more characters
print(txt.find('w', 3, -5)) # first occurance of character from index 3 to last 6th (last 5 excluded)
print(txt.find('!', 2, -1)) # first occurance of character from index 2 to last 2nd (last one excluded)
```

```python
# index of one or more characters of first occurance (ValueError if no occurance)
# str.index(sub: str, start: SupportsIndex | None = None, end: SupportsIndex | None = None,) -> int
# sub: str
# start: SupportsIndex -> int, def, bool -> True = 1, False = 0
# end: SupportsIndex -> int, def, bool -> True = 1, False = 0
txt = 'hello world!'
print(txt.index('hello', False)) # index of first occurance from 0
print(txt.index('d', 7, -1)) # first occurance of character from index 2 to last 2nd (last one excluded)
```

```python
# check the string is alpha or numeric or alpha-numaric and at least one character in string
# a-z or 0-9 or a-z0-9
print('hello!'.isalnum()) # contain '!' spacial character = false
print('hello'.isalnum())
print('123424'.isalnum())
print('hello 25'.isalnum()) # contain ' ' spacial character = false
print('hello25'.isalnum())
```

```python
# check the string is only alpha characters and at least one character in string
# a-z or A-Z or a-zA-z
print('hello'.isalpha())
print('WORLD'.isalpha())
print('HelloWorld'.isalpha())
print('123424'.isalpha())
print('hello 25'.isalpha())
print('hello25'.isalpha())
```

```python
# check the string is only numeric characters and at leat one character in string
# 0 - 9 
print('1234567890'.isdecimal())
print('abcd12'.isdecimal())
print('world 26'.isdecimal())
print('12345!'.isdecimal())
print('1'.isdecimal())
```

```python
# check the string is ASCII characters and at least one character in string
# a-z,A-Z,0-9, all the keyboard characters
print('12345'.isascii())
print('abcdefghijklmnopqrstuvwxyz 1234567890 !@#$%^&*()_-+=~`/?.<,>'.isascii())
print('😂, 🚀, 👍'.isascii()) # false not a ascii character
```

```python
# check the string is numeric and at least one character in string
# 0-9 and unicode numeric characters
print('0987654321'.isnumeric())
print('\u2164'.isnumeric())
print('十'.isnumeric())
```

```python
# check the string is python veriable idendifier
# start with alpha(a-z or A-Z) or _ characters => True, otherwise False
# using '-' => False
print('myname'.isidentifier())
print('my_name'.isidentifier())
print('my_Name'.isidentifier())
print('My_Name'.isidentifier())
print('myName'.isidentifier())
print('my-name'.isidentifier())
print('_myname'.isidentifier())
print('-myname'.isidentifier())
print('MyName'.isidentifier())
print('myname24'.isidentifier())
print('24myname'.isidentifier())
```

```python
# check the string is lower and at one lower cased character in string
# a-z
print('hello'.islower())
print('hello '.islower())
print('abcdefghijklmnopqrstuvwxyz 1234567890 !@#$%^&*()_-+=~`/?.<,>'.islower())
print('hellO'.islower())
```

```python
# check the string is space and at least one space in string
print(' '.isspace())
print('    '.isspace())
print('asf  '.isspace())
```

```python
# check the string every word first letter is capital letter
print('Hello Word'.istitle())
print('hello World'.istitle())
print('Hello world'.istitle())
```

```python
# check the string is upper cased
# A-Z and keyboard characters
print('HELLO'.isupper())
print('HELLO 1234567890 !@#$%^&*()_-+=~`/?.<,>'.isupper())
print('Hello'.isupper())
```

```python
# remove all the characters from both ends
# str.strip(chars: LiteralString | None = None)
# chars: LiteralString -> (default is space(' ')) single or more characters
print('    hello world!    '.strip())
print('cadbhello world!dcab'.strip('bca')) # (match all the characters in shuffled also valid (abcd, bcda, dbac, ...))
```
