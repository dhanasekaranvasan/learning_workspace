Set is unorder, remove duplicates, hashable, mutable
Types: str, int, float, bool, tuple, frozenset, def

```python
nums = {1,2,3,4,5,6,1,"2",3,4,5,6,1,2,3,4,5,6}
names = {"name1", "name2", "name3", "name4", "name5", "name6"}
bools = {0,"True",0.0,False, "False", True, "0", 1, "1", "0.0"}
print(nums)
print(names)
print(bools)
```

```python
# add one new element
print(nums)
nums.add('new_item')
print(nums)
```

```python
# update element 
# str, tuple
names.update('welcome') # update string iterate every character
print(names)
names.update(('again1', 'again2')) # update tuple iteate very item
print(names)
```

```python
# update set by iterate elements
bools.update("string") # iterate every string characters
print(bools)
bools.update((1,2,4,5))
print(bools)
```

```python
# check the set is subset
subset = {2,4}
print(subset.issubset(nums))
print({'welcome'}.issubset(names))
```

```python
# find the difference in set
print(nums.difference({1,4,5}))
print(names.difference({'name1', 'name2'}))
```

```python
# remove item from set (ValueError if not found)
print(nums)
nums.remove(1)
print(nums)
```

```python
# discard element from set (no error)
print(bools)
bools.discard('True')
print(bools)
```

```python
# union, get exists element and add element if not exist
print(nums.union({'10', 20, 'added', 1,2,3,4}))
```

```python
# match all the elements is superset(True) (Otherwise False)
print({1, 2, 3, 4, 5, 6, '2'}.issuperset(nums))
print({'names'}.issuperset(names))
```
