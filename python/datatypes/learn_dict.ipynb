Dict is insertion order, mutable, key value pairs, key dont allow duplicate, allow duplicate values
Key types: str, int,float, bool, tuple, frozenset
Value types: str, int, float, bool, tuple, frozenset, list, set, dict

```python
collections = {
    'nums': [1, 2, 3, 4, 5, 6], 
    'points': (1.4, 5.0, 7.8, 324.3, 12.5, 23.7),
    'names': {'Kumar', 'Arun', 'Vicky', 'Vishal', 'Selvi', 'Guru'},
    'info': {
        'nums': [1, 2, 3, 4, 5, 6], 
        'points': (1.4, 5.0, 7.8, 324.3, 12.5, 23.7),
        'names': {'Kumar', 'Arun', 'Vicky', 'Vishal', 'Selvi', 'Guru'},
    }
}
print(collections)
```

```python
# get all items with key and value
print(collections.items().mapping)
```

```python
# get keys
print(collections.keys().mapping)
```

```python
# get values
print(collections.values().mapping)
```

```python
# get the values my key (None if key not found)
print(collections.get("nums"))
```

```python
# get the value by key (KeyError if key not found)
print(collections['nums'])
```

```python
# update item by key (new key created by value, exist key updated by new value)
old = collections['nums']
collections['nums'] = "new updated string value"
print(collections)
collections['nums'] = old
print(collections)
collections.update({'nums': ["test"]})
print(collections)
collections.update({'nums': old, (1,3): 'tuple based item'})
print(collections)
```

```python
# remove item by key (KeyError if not found)
collections.pop('nums')
print(collections)
```
