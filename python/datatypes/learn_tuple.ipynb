Tuple is ordered, immutable, allow duplicates
Types: Types: str, int, float, bool, dict, set, tuple, def, object

```python
# (,) -> tuple
# ('str') -> not tuple
# (1) -> not tuple
# ([]) -> not tuple
# ((1)) -> not tuple
# ({}) -> not tuple
nums = (1,2,3,4,5,6,1,2,3,4,5,6,1,2,3,4,5,6)
names = ("name1", "name2", "name3", "name4", "name5", "name6")
types = (nums, names, ["list1", nums, names], (nums, names), {nums, names}, {'nums': nums, 'names': names})
print(nums)
print(names)
print(types)
```

```python
# count occurance of elements
print(nums.count(2))
```

```python
# find the first occurance of element (ValueError if not found)
print(names.index("name2"))
print(names.index("name5", 2, -1)) # find the element from index 2 to last 2nd (last one excluded)
```
