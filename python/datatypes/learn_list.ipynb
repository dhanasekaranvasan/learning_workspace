List is order, allow duplicates, mutable
Types: str, int, float, bool, dict, set, tuple, def, object

```python
names = ["Arun", "Kumar", "Selvi", "Balu", "Dhana", "Sekar", "Karthik", "Rajesh", "Kannan", "Vasan", "Malliga", "Karthik", [1,2]]
def func(names):
    indices = [r for r in range(0, len(names))]
    print(indices)
    print(names)
func(names)
```

```python
# find the elements bt index (IndexError out of range, if no index found)
func(names)
print('5th index:', names[5]) # find the element at index 5
# print('11th index:', names[11]) # find the element at index 11 (IndexError)
```

```python
# append one or more elements at end of list (no iterate, only append elements by specified type)
# str, int, float, bool, list, tuple, set, dict, object
print("Before:")
func(names)
names.append(["Arjun","Josh"])
print("After:")
func(names)
names.append([1, "2", "abcd", True, False, ("Hello",), {"eat", "World"}, {'name': 'Arul', 'age': 23}])
func(names)
names.append(("tuple1", "tuple2", 125, ("tuple", "tuple_inside"), {"set1", "set2"}, ["list1", "list2"], {'name': 'Arul', 'age': 23}))
func(names)
```

```python
# extend one or more elements at end of list (extend elements by iterate)
# str, int, float, bool, list, tuple, set, dict, object
print("Before:")
func(names)
print("After:")
names.extend(["Arjun","Josh"])
func(names)
names.extend("String") # string iterate every single character as a element
func(names)
names.extend([1, "2", "abcd", True, False, ("Hello",), {"eat", "World"}, {'name': 'Arul', 'age': 23}])
func(names)
names.extend(("tuple1", "tuple2", 125, ("tuple", "tuple_inside"), {"set1", "set2"}, ["list1", "list2"], {'name': 'Arul', 'age': 23}))
func(names)
```

```python
# insert element by index
# supprtsindex -> int, bool -> (True = 1, False = 0)
# operation same us append
print("Before:")
func(names)
print("After:")
names.insert(5, "Index 5")
names.insert(2, 1234)
names.insert(5, True)
names.insert(0, ["Index 5", 12345, True, ("tuple1",)])
func(names)
```

```python
# update the elemets by index (replace exist element, IndexError out of range if index not found)
# slicing index same us extand and insert
print("Before:")
func(names)
print("After:")
names[0] = "updated index 0" # update element at index 0
func(names)
names[-1] = "last element" # update element at index -1 (last element)
func(names)
names[2:6] = "substring" # update and insert element at index 2 to 5th (6th excluded) and string iterate every characters
func(names)
names[2:11] = ["2234", "3", 4 ,5,6] # update and insert element at index 2 to 10th (11th excluded)
func(names)
```

```python
# sorting the elements based on key (all elements are same type, otherwise TypeError)
values = ["text", "Text", 1, "1", "0", "True", False, True, "False", "0.0", 0, 0.0, [2,4,5], (3,4,5)]
func(values)
# values.sort() # TypeError multiple type not supported
values.sort(key=str)
func(values)
```

```python
# remove element by index (default is -1)(IndexError out of range, if index not found)
print("Before:")
func(names)
print("After:")
names.pop() # remove the last index (-1)
func(names)
names.pop(3) # remove the 3rd index
func(names)
names.pop(-2) # remove the last 2nd index
func(names)
# names.pop(17) # remove the 17th index (IndexError)
func(names)
```

```python
# remove the first occurance of element by value (IndexError if not found)
print("Before:")
func(names)
print("After:")
names.remove("Karthik")
func(names)
names.remove([1,2])
func(names)
```

```python
# copy the list
new_names = names # bother are same, operations also same
func(names)
func(new_names)
new_names.append("updated")
func(names)
func(new_names)
new_names = names.copy() # both are different list
func(names)
func(new_names)
new_names.extend("extended")
func(names)
func(new_names)
```
